# What are the non-major related job percentages for each major?
<hr/>

Getting rid of the Major Military Technologies because it has 0s

```python
no_mil = recent_grads.drop(index = 73)
```

Taking the percentage of non college jobs in the entire amount of people in major

```python
no_mil["Job_percentage"] = no_mil["Non_college_jobs"]/(no_mil["College_jobs"] + no_mil["Non_college_jobs"])
```

Creating a view that only shows a few columns and sorting it in descending values

```python
no_mils = no_mil.loc[:, ['Rank', 'Major', 'College_jobs', "Non_college_jobs", "Job_percentage"]]
jobsratio = no_mils.sort_values(by=['Job_percentage'], ascending = False)
```

Final formatting of information

```python
for major,jobp in zip(jobsratio.Major, jobsratio.Job_percentage):
    print("The Major", major, "has a percentage of people who ended up a job unrelated to their major at", "{:.0%}".format(jobp) + ".")    
```
