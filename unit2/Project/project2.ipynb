Which jobs have higher percentages of women than men? 

```python
#gender = recent_grads.loc[:, ['Rank', 'Major', 'Men', "Women", "Total", "ShareWomen", ]]
gender = recent_grads.loc[(recent_grads.Women >= recent_grads.Men), ['Rank', 'Major', 'Men', "Women", "Total", "ShareWomen", ]]

genders = gender.sort_values(by=['ShareWomen'], ascending = False)
genders
#genders
```

```python
temp_genders = [print("A major that has a higher percentage of women than men is", genderss, "at", "{:.0%}".format(genderssss), "women."
) for genderss,genderssss in zip(genders.Major, genders.ShareWomen)]
```

```python
print("There are", len(temp_genders), "majors that have a larger percentage of women than men.")
```

```python
recent_grads.loc[recent_grads.Major_category == "Biology & Life Science", ["Major", "Total", "Men", "Women", "ShareWomen"]]
```

```python
#What majors have under 10 percent women
sad_gender = gender = recent_grads.loc[(recent_grads.ShareWomen <= 0.20), ['Rank', 'Major', 'Men', "Women", "Total", "ShareWomen", ]]
sad_genders = sad_gender.sort_values(by=['ShareWomen'], ascending = True)

temp_sad_genders = [print("A major that has under 20% of women than men is", genderss, "at", "{:.0%}".format(genderssss), "women."
) for genderss,genderssss in zip(sad_genders.Major, sad_genders.ShareWomen)]
temp_sad_genders
```
