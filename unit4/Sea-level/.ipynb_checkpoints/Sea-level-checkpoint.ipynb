```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#lvls is sea levels
lvls = pd.read_csv("epa-sea-level.csv")
lvls.drop(["Lower Error Bound", "Upper Error Bound", "NOAA Adjusted Sea Level"], axis = "columns", inplace = True)
lvls.rename({"CSIRO Adjusted Sea Level": 'sea-level'}, axis = 'columns', inplace = True)
lvls.info()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.regplot(data = lvls, x = "Year", y = "sea-level")
```

```python
X = lvls.Year.values.reshape(-1,1)
y = lvls["sea-level"].values.reshape(-1,1)

model = LinearRegression().fit(X,y)

model.score(X,y)
```

```python
year = 2070
y_pred = model.predict([[year]])
print("The sea-level in year", str(year), "is predicted to be: {:.2f}".format(y_pred[0][0]), "inches.")
```

```python
diff = y_pred[0][0] - lvls.loc[lvls["Year"] == 2012, "sea-level"].values[0]
print(f'In {year} the sea level will be {round(diff,2)} inches higher than in {2012}.')
```
