# One Line Version

```python
["fizzbuzz" if x%3 == 0 and x%5 == 0 else "fizz" if x%3 == 0 else "buzz" if x%5 == 0 else x for x in range(100) ]
```

# Many Line Version

```python
for x in range(100):
    if x%3 == 0 and x%5 == 0:
        print("fizzbuzz")
    elif x%5 == 0:
        print("buzz")
    elif x%3 == 0:
        print("fizz")
    else:
        print(x)
```
